```python
from atmons import *
import matplotlib.pyplot as plt
```

```python
config, grid = loader()
# config['spec_key'] = 'be'
config['spec_key'] = 'wfc'
config['chem'] = 's001'

config['m_ns'] = 1.5
config['r_ns'] = 13.16
config['v_rot'] = 700

grid['n_nu'] = 500
grid['n_theta'] = 6
grid['n_phi'] = 12
grid['rng_erg'] = [3.0, 20.0]

dumper('lfc', config, grid)
config, grid = loader("lfc")
```

```python
SizeA = GraphSize(x_range=(0.0,1.1), y_range=(1.3,1.9),  figsize=(7,6))
SizeB = GraphSize(x_range=(0.0,1.1), y_range=(0.07,0.27), figsize=(7,6))
I_arr = [0, 45, 90]
Fkey_arr = [1, 2]
for k in range(len(Fkey_arr)):
    config['flux_key'] = Fkey_arr[k]
    flux_key = config['flux_key'] 
    for i in range(len(I_arr)):
        config['i_ang'] = I_arr[i]
        W_arr, Fc_arr, L_arr = [], [], []
        ns = NeurtonStar(config, grid)
        burst = ns.burst()
        counter = 0
        for shot in burst:
            #print(f"{counter} | {shot.lum}")
            L_arr.append(shot.lum)
            Fc_arr.append(shot.fc)
            W_arr.append(shot.w)
            counter += 1
            percent = (k*3*N_MODEL+i*N_MODEL+(counter))/(2*3*N_MODEL) * 100
            print(f"Downloading...{percent:.2f}%: {k+1} graph, {i+1} line, {counter} shot ")

        data_w = GraphData(L_arr, W_arr, tex_mode=True,
                            ascii_label=f"i = {I_arr[i]} deg",
                            latex_label=f"$i = {I_arr[i]}, deg$")
        data_fc = GraphData(L_arr, Fc_arr, tex_mode=True,
                            ascii_label=f"i = {I_arr[i]} deg",
                            latex_label=f"$i = {I_arr[i]}, deg$")
        if flux_key==1:
            DataBaseFig8A.DB.append(data_fc)
            DataBaseFig8B.DB.append(data_w)
        else:
            DataBaseFig9A.DB.append(data_fc)
            DataBaseFig9B.DB.append(data_w)
    
    if flux_key == 1:
        Graph_8A = GraphOld(DataBaseFig8A, SizeA, LegendStyle)
        Graph_8B = GraphOld(DataBaseFig8B, SizeB, LegendStyle)
        Graph_8A.draw_picture(save=True, draw=False) 
        Graph_8B.draw_picture(save=True, draw=False) 
    else:
        Graph_9A = GraphOld(DataBaseFig9A, SizeA, LegendStyle)
        Graph_9B = GraphOld(DataBaseFig9B, SizeB, LegendStyle)
        Graph_9A.draw_picture(save=True, draw=False) 
        Graph_9B.draw_picture(save=True, draw=False) 

Graph_arr = [Graph_8A, Graph_8B, Graph_9A, Graph_9B]
for graph in Graph_arr:
    graph.draw_picture(save=True, draw=False) 
    graph.draw_table() 
```
